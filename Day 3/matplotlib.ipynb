```python
import matplotlib.pyplot as plt
%matplotlib inline

#using ""%matplotlib inline" will let us use only plt instead of plt.show
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
a = np.arange(40,50)
b = np.arange(51,60)
```

```python
#plot using matplotlib

plt.scatter(x,y,color="r")

#name the x and y axes labes
plt.xlabel('Age')
plt.ylabel('Number')

#give your plot a title
plt.title('Age-Num plot')
plt.savefig('plot.png')
```

```python
#create subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r')

plt.subplot(2,2,2)
plt.plot(x,y,'c')

plt.subplot(2,2,3)
plt.plot(x,y,'g*')
```

```python
x = np.arange(1,10)
y = np.sin(x)
plt.plot(x,y,)
plt.xlabel('Num')
plt.ylabel('Quantity')
plt.title('Amount graph')
```

```python
#bar plots

x = np.arange(1,10)
y = np.sin(x)
plt.bar(x,y,)
plt.xlabel('Num')
plt.ylabel('Quantity')
plt.title('Amount graph')
```

```python
#histogram

x = ([2,6,8,12,19,35])
plt.hist(x)
plt.xlabel('Num')
plt.ylabel('Quantity')
plt.title('Amount graph')
```
